```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
PATH = os.getcwd()
DATAPATH = os.path.join(PATH, "data")
filenames = {
    "X_test": "test_in - Copy.csv",
    "X_train": "train_in - Copy.csv",
    "y_test": "test_out - Copy.csv",
    "y_train": "train_out - Copy.csv",
}

# Import all data files
X_train = pd.read_csv(os.path.join(DATAPATH, filenames["X_train"]), header=None)
y_train = pd.read_csv(
    os.path.join(DATAPATH, filenames["y_train"]), header=None, names=["digit"]
)
X_test = pd.read_csv(os.path.join(DATAPATH, filenames["X_test"]), header=None)
y_test = pd.read_csv(
    os.path.join(DATAPATH, filenames["y_test"]), header=None, names=["digit"]
)
```

```python
def sigmoid(x,a=1):
    return 1/(1 + np.exp(-a*x))

def dsigmoid(x,a=1):
    return sigmoid(x,a)(1-sigmoid(x,a))

def err(y,d):
    return 1/2*np.sum(np.square(y-d))
```

```python
X_train['Bias'] = np.ones((len(X_train),))
X_test['Bias'] = np.ones((len(X_test),))

def rand_weights():
    return pd.DataFrame(np.random.uniform(-1,1,(257,10)))

weight_train = rand_weights()
activations = np.dot(X_train,weight_train)
activations_sigmoid = pd.DataFrame(sigmoid(activations))
y_pred = [np.argmax(row) for _,row in activations_sigmoid.iterrows()]
yreal = np.zeros((len(X_train),10))
for i in range(len(X_train)):
    j = y_train.digit[i]
    yreal[i,j] = 1

err(activations_sigmoid.iloc[0,:],yreal[0,:])
 
pd.DataFrame(activations_sigmoid) 
```
