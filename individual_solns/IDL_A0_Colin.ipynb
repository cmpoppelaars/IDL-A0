```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = os.getcwd()
DATAPATH = os.path.join(PATH, "../data")

filenames = {
    "X_test":'test_in - Copy.csv',
    "X_train":'train_in - Copy.csv',
    "Y_test":'test_out - Copy.csv',
    "Y_train":'train_out - Copy.csv'
    
}
```

```python
X_test, y_test = pd.read_csv(os.path.join(DATAPATH, filenames["X_test"])), pd.read_csv(os.path.join(DATAPATH, filenames["Y_test"]))
X_train, y_train = pd.read_csv(os.path.join(DATAPATH, filenames["X_train"])), pd.read_csv(os.path.join(DATAPATH, filenames["Y_train"]))
```

```python
# Filter all the zeros with the zeros, etc. 

## Define the range of numbers from 0 to 9
numbers_to_filter = np.arange(10)

## Create a boolean mask for each number
masks = [(y_train == number) for number in numbers_to_filter]

## Use np.where to get indices for each mask
indices_of_numbers = [np.where(mask)[0] for mask in masks]

# Example: 0
## all indices with the number 0
print(indices_of_numbers[0])
## all zeros by applying the filter to df
print(y_train.iloc[indices_of_numbers[0]])
```

```python
# Getting the mean of the all the 0's
print(X_train.iloc[indices_of_numbers[0]].mean())
```

```python
def avg_vec(number):
    return X_train.iloc[indices_of_numbers[number]].mean()
    
# All number averages  
centre = np.array([avg_vec(i) for i in range(10)])
for i in centre:
    plt.imshow(i.reshape((16,16)))
    plt.show()
```

```python
# Adding a distance measure
def d(cent, data):
    return sum(((data-cent))**2)


distances = [[d(i, j) for i in X_train] for j in centre]
```
